```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2

csv = pd.read_csv("onlinefoods2.csv")

def limpiarColumnaGenero(valor):
  valor = str(valor)
  valor = valor.replace("Male", "M")
  valor = valor.replace("Female", "F")
  return valor


def limpiarColumnaStatus(valor):
  valor = str(valor)
  valor = valor.replace("Single", "S")
  valor = valor.replace("Married", "M")
  if valor == "Prefer not to say":
    valor = np.nan
  return  valor


def limpiarColumnaEdcucacional(valor):
  valor = str(valor)
  if valor == "Ph.D":
    valor = np.nan
  return  valor


csv['Gender'] = csv['Gender'].apply(limpiarColumnaGenero)
csv['Marital Status'] = csv['Marital Status'].apply(limpiarColumnaStatus)
csv['Educational Qualifications'] = csv['Educational Qualifications'].apply(limpiarColumnaEdcucacional)
csv = csv.dropna()



labels = csv['Educational Qualifications'].unique()
sizes = [(csv['Educational Qualifications'] == labels[0]).sum(), (csv['Educational Qualifications'] == labels[1]).sum(), (csv['Educational Qualifications'] == labels[2]).sum()]
explode = (0, 0, 0)
fig1, ax1 = plt1.subplots(figsize=(5, 5))
ax1.pie(sizes, labels=labels, explode=explode, autopct='%.2f%%',startangle=120, pctdistance=0.5)
plt1.title('Porcentaje de los tipos de preparaciones acadmicas de los clientes:')
plt1.show()



plt1.bar(csv['Gender'].unique(),[(csv['Educational Qualifications'] == labels[0]).sum(), (csv['Educational Qualifications'] == labels[1]).sum()])
plt1.title('Gnero de los clientes:')
plt1.show()



array = sorted(csv["Age"].unique(), reverse=True)
array2 = []
for i in array:
  array2.append((csv['Age'] == i).sum())
plt1.plot(array, array2)
plt1.title("Edad de los clientes")
plt1.xlabel('Edad')
plt2.ylabel('Cantidad')
plt1.show()
```

La mayora de los clientes tienen entre 22 y 24 aos, de gnero masculino y estudiaron un posgrado.
