```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv = pd.read_csv("Formulario.csv")

def simplificarFisicoDigital(valor):
  valor = str(valor)
  valor = valor.replace("Fsico", "F")
  valor = valor.replace("Digital", "D")
  return valor

def cambiarAOtro(valor):
  valor = str(valor)
  if valor != "Comodidad" and valor != "Portabilidad" and valor!= "Disponibilidad / Variedad de ttulos" and valor != "Disponibilidad de idioma"and valor != "Espacio" and valor != "Sostenibilidad ambiental" and valor !="Costo" and valor != "Coleccionismo" and valor!= "Anotaciones" and valor != "Esttica":
    valor.replace(valor, "Otro")
  return valor

csv['En qu formato suele leer ms?'] = csv['En qu formato suele leer ms?'].apply(simplificarFisicoDigital)
csv['En qu formato prefiere leer?'] = csv['En qu formato prefiere leer?'].apply(simplificarFisicoDigital)
csv['Cul es la principal razn por la que prefiere leer en el formato seleccionado?'] = csv['Cul es la principal razn por la que prefiere leer en el formato seleccionado?'].apply(cambiarAOtro)

nombres = csv['Qu edad tiene?'].unique()
valores = []
for valor in nombres:
  valores.append((csv['Qu edad tiene?'] == valor).sum())
print(valores)
plt.bar(nombres, valores)
plt.show()

filas_con_digital = csv[csv['En qu formato suele leer ms?'] == 'D']
filas_fatiga_visual = filas_con_digital[filas_con_digital['Le provoca ms fatiga visual leer en digital que en fsico?'] == 'S']
nombres = ["S", "No"]
valores = []
for valor in nombres:
  valores.append((filas_con_digital['Le provoca ms fatiga visual leer en digital que en fsico?'] == valor).sum())
explode = (0, 0)
fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(valores, labels=nombres, explode=explode, autopct='%.2f%%',startangle=120, pctdistance=0.7)
plt.title('Porcentaje de personas que suelen leer en DIGITAL y les provoca ms fatiga visual leer en digital que en fsico:')
plt.show()

filas_con_fisico = csv[csv['En qu formato suele leer ms?'] == 'F']
nombres = ["S", "No"]
valores = []
for valor in nombres:
  valores.append((filas_con_fisico['Le provoca ms fatiga visual leer en digital que en fsico?'] == valor).sum())
explode = (0, 0)
fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(valores, labels=nombres, explode=explode, autopct='%.2f%%',startangle=120, pctdistance=0.7)
plt.title('Porcentaje de personas que suelen leer en FSICO y les provoca ms fatiga visual leer en digital que en fsico:')
plt.show()

filas_con_costo = csv[csv['Ha considerado leer un libro en su versin digital debido a su costo en fsico?'] == 'S']
nombres = ['Todos los das', 'Varias veces a la semana', 'Una vez a la semana', 'Varias veces al mes', 'Una vez al mes', 'Cada 3 meses', 'Cada 6 meses', 'Nunca']
valores = []
for valor in nombres:
  valores.append((filas_con_costo['Con qu frecuencia lee libros?'] == valor).sum())
plt.figure(figsize=(10, 6))
plt.bar(nombres, valores)
plt.xticks(rotation=45, ha='right')
plt.title('Con cunta frecuencia leen las personas que consideraron leer un libro en su versin digital debido a su costo en fsico:')
plt.show()

filas_sin_costo = csv[csv['Ha considerado leer un libro en su versin digital debido a su costo en fsico?'] == 'No']
nombres = ['Todos los das', 'Varias veces a la semana', 'Una vez a la semana', 'Varias veces al mes', 'Una vez al mes', 'Cada 3 meses', 'Cada 6 meses', 'Nunca']
valores = []
for valor in nombres:
  valores.append((filas_sin_costo['Con qu frecuencia lee libros?'] == valor).sum())
plt.figure(figsize=(10, 6))
plt.bar(nombres, valores)
plt.xticks(rotation=45, ha='right')
plt.title('Con cunta frecuencia leen las personas que NO consideraron leer un libro en su versin digital debido a su costo en fsico:')
plt.show()



```
