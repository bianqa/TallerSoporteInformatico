```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv = pd.read_csv("Formulario.csv")

def simplificarFisicoDigital(valor):
  valor = str(valor)
  valor = valor.replace("Físico", "F")
  valor = valor.replace("Digital", "D")
  return valor

def cambiarAOtro(valor):
  valor = str(valor)
  if valor != "Comodidad" and valor != "Portabilidad" and valor!= "Disponibilidad / Variedad de títulos" and valor != "Disponibilidad de idioma"and valor != "Espacio" and valor != "Sostenibilidad ambiental" and valor !="Costo" and valor != "Coleccionismo" and valor!= "Anotaciones" and valor != "Estética":
    valor.replace(valor, "Otro")
  return valor

csv['¿En qué formato suele leer más?'] = csv['¿En qué formato suele leer más?'].apply(simplificarFisicoDigital)
csv['¿En qué formato prefiere leer?'] = csv['¿En qué formato prefiere leer?'].apply(simplificarFisicoDigital)
csv['¿Cuál es la principal razón por la que prefiere leer en el formato seleccionado?'] = csv['¿Cuál es la principal razón por la que prefiere leer en el formato seleccionado?'].apply(cambiarAOtro)

nombres = csv['¿Qué edad tiene?'].unique()
valores = []
for valor in nombres:
  valores.append((csv['¿Qué edad tiene?'] == valor).sum())
print(valores)
plt.bar(nombres, valores)
plt.show()

filas_con_digital = csv[csv['¿En qué formato suele leer más?'] == 'D']
filas_fatiga_visual = filas_con_digital[filas_con_digital['¿Le provoca más fatiga visual leer en digital que en físico?'] == 'Sí']
nombres = ["Sí", "No"]
valores = []
for valor in nombres:
  valores.append((filas_con_digital['¿Le provoca más fatiga visual leer en digital que en físico?'] == valor).sum())
explode = (0, 0)
fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(valores, labels=nombres, explode=explode, autopct='%.2f%%',startangle=120, pctdistance=0.7)
plt.title('Porcentaje de personas que suelen leer en DIGITAL y les provoca más fatiga visual leer en digital que en físico:')
plt.show()

filas_con_fisico = csv[csv['¿En qué formato suele leer más?'] == 'F']
nombres = ["Sí", "No"]
valores = []
for valor in nombres:
  valores.append((filas_con_fisico['¿Le provoca más fatiga visual leer en digital que en físico?'] == valor).sum())
explode = (0, 0)
fig1, ax1 = plt.subplots(figsize=(5, 5))
ax1.pie(valores, labels=nombres, explode=explode, autopct='%.2f%%',startangle=120, pctdistance=0.7)
plt.title('Porcentaje de personas que suelen leer en FÍSICO y les provoca más fatiga visual leer en digital que en físico:')
plt.show()

filas_con_costo = csv[csv['¿Ha considerado leer un libro en su versión digital debido a su costo en físico?'] == 'Sí']
nombres = ['Todos los días', 'Varias veces a la semana', 'Una vez a la semana', 'Varias veces al mes', 'Una vez al mes', 'Cada 3 meses', 'Cada 6 meses', 'Nunca']
valores = []
for valor in nombres:
  valores.append((filas_con_costo['¿Con qué frecuencia lee libros?'] == valor).sum())
plt.figure(figsize=(10, 6))
plt.bar(nombres, valores)
plt.xticks(rotation=45, ha='right')
plt.title('Con cuánta frecuencia leen las personas que consideraron leer un libro en su versión digital debido a su costo en físico:')
plt.show()

filas_sin_costo = csv[csv['¿Ha considerado leer un libro en su versión digital debido a su costo en físico?'] == 'No']
nombres = ['Todos los días', 'Varias veces a la semana', 'Una vez a la semana', 'Varias veces al mes', 'Una vez al mes', 'Cada 3 meses', 'Cada 6 meses', 'Nunca']
valores = []
for valor in nombres:
  valores.append((filas_sin_costo['¿Con qué frecuencia lee libros?'] == valor).sum())
plt.figure(figsize=(10, 6))
plt.bar(nombres, valores)
plt.xticks(rotation=45, ha='right')
plt.title('Con cuánta frecuencia leen las personas que NO consideraron leer un libro en su versión digital debido a su costo en físico:')
plt.show()



```
